## 1. Load Data

```python
import pandas as pd

credit_df = pd.read_csv("../data/raw/creditcard.csv")
credit_df.head()
```

## 2. Understand Dataset

```python
# Basic info
credit_df.info()

# Check missing values
credit_df.isnull().sum()

# Class distribution
credit_df['Class'].value_counts(normalize=True)
```

## 3. Exploratory Data Analysis
### a) Target distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='Class', data=credit_df)
plt.title('Fraud vs Non-Fraud Transactions')
plt.show()
```

## b) Transaction amount distribution

```python
sns.histplot(credit_df['Amount'], bins=50, kde=True)
plt.title('Transaction Amount Distribution')
plt.show()
```

## c) Time variable

```python
sns.histplot(credit_df['Time'], bins=50, kde=True)
plt.title('Time Distribution of Transactions')
plt.show()
```

## d) Correlation heatmap

```python
plt.figure(figsize=(12,10))
sns.heatmap(credit_df.corr(), cmap='coolwarm', center=0)
plt.title('Feature Correlation')
plt.show()
```

## 5. Preprocessing for Modeling

### Scale Amount and Time.

### Keep PCA features as-is.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
credit_df['Amount_scaled'] = scaler.fit_transform(credit_df[['Amount']])
credit_df['Time_scaled'] = scaler.fit_transform(credit_df[['Time']])
```
